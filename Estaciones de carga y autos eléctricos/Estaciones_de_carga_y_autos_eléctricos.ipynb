```python
# Construir el grafo dual
mapa = nx.Graph()

# Obtener las caras del grafo planar
caras = list(nx.cycle_basis(Ga))

# Agregar las caras como nodos al grafo dual
for i, cara in enumerate(caras):
    mapa.add_node(i + 1)  # El nodo i + 1 representa la cara i+1

# Agregar aristas al grafo dual según las aristas compartidas entre las caras
for i, cara1 in enumerate(caras):
    for j, cara2 in enumerate(caras[i+1:]):
        for nodo in cara1:
            if nodo in cara2:
                mapa.add_edge(i + 1, i + j + 2)  # Las caras i+1 y i+j+2 están conectadas

# Generar la disposición plana del grafo
pos = nx.planar_layout(mapa)

# Ajustar el factor de escala del largo de las aristas
edge_scale_factor = 2

# Escalar las posiciones de los nodos para ajustar el largo de las aristas
scaled_pos = {node: (x * edge_scale_factor, y * edge_scale_factor) for node, (x, y) in pos.items()}

# Dibujar el grafo
nx.draw(mapa, scaled_pos, with_labels=True, node_color='lightblue', node_size=325, font_size=7, edge_color='gray', width = 2)

plt.show()
```

```python
class auto():
    def __init__(self,placa, bateria, ubicacion):
        self.placa = placa
        self.bateria = bateria
        self.ubicacion = ubicacion

    def __repr__(self):
         return "({0},{1},{2})".format(self.placa, self.bateria, self.ubicacion)

    def AOC(self, n, E):
      Caminos = []
      CaminosM = []
      spc = []
      E_nodos = [x.ubicacion for x in E.total]
      if self.ubicacion in E_nodos:
        return self.ubicacion
      else:
        for i in range(0,n):
          Ch = []
          eCh = 0
          P = []
          Ch.append(self.ubicacion)
          while Ch[-1] not in E_nodos:
            l = []
            for a in list(G.neighbors(Ch[-1])):
              if a not in Ch:
                l.append(a)
            if len(l)>0:
              j = random.choice(l)
              dab = G[Ch[-1]][j]['weight']
              Ch.append(j)
              P.append(1/dab)
              a = j
              eCh += 1
            else:
              CaminosM.append(Ch)
          Caminos.append(Ch)
          sp = sum(P)*(100/eCh)
          spc.append(sp)
        indice = spc.index(max(spc))
        return Caminos[indice]

class flota():
    def __init__(self, total):
        f = []
        for i in range(total):
            placa = "car"+str(i)
            bateria = random.randrange(50,100)
            ubicacion = random.choice(list(G.nodes))
            autoi = auto(placa, bateria, ubicacion)
            f.append(autoi)
        self.total = f

    def __repr__(self):
        pf = ""
        for auto in self.total:
            pf += repr(auto)
        return pf

class enrutador():
    def __init__(self, enrutamiento, nodo):
        self.enrutamiento = enrutamiento
        self.nodo = nodo

    def nodos_adyacentes(self,r, nodo):
      l = []
      for i in list(Gr.neighbors(nodo)):
        if i not in r and len(list(Gr.neighbors(nodo))) != 0:
          l.append(i)
          return l

class servidor(enrutador):
    def __init__(self, ip, dtb,nodo, enrutamiento):
        self.ip = ip
        self.dtb = dtb
        self.nodo = nodo
        self.enrutamiento = enrutamiento

    def __repr__(self):
      return "ip: {0}, dtb: {1}".format(self.ip, self.dtb)

class antenas(enrutador):
    def __init__(self):
      t = []
      s = 0
      enr = {}
      for i in Ga:
        enr[i]=[i,float("inf")]
        antena = servidor(s, [], i,enr)
        s+=1
        t.append(antena)
      self.total = t

    def __repr__(self):
      r = ""
      for i in self.total:
        r += repr(i)+"\n"
      return r


class estacion_carga(enrutador):
    def __init__(self, capacidad, cola, slots, ubicacion, nodo, ip, enrutamiento):
        self.cola = cola
        self.slots = slots
        self.ubicacion = ubicacion
        self.nodo = nodo
        self.ip = ip
        self.capacidad = capacidad
        self.enrutamiento = enrutamiento

    def __repr__(self):
       return "ip: {0}, nodo: {2}, Ubicacion: {1},  Slots disponibles: {3}, Autos en cola: {4}".format(self.ip, self.ubicacion, self.nodo, self.capacidad-len(self.slots),len(self.cola),)

class estaciones(estacion_carga):
    def __init__(self):
      t = []
      s = 1
      enr = {}
      for i in Gr:
        enr[i]=[i,float("inf")]
        estacion_c = estacion_carga(10, [], [], random.choice(ubicaciones[i]),i,s,enr)
        t.append(estacion_c)
        s += 1
      self.total = t

    def __repr__(self):
      m = ""
      for i in self.total:
        m += repr(i)+"\n"
      return m

```

```python

def dijkstra1(Grafo, salida):
    dist, prev = {}, {} #diccionarios para la distancia desde el nodo origen y nodo previo en la ruta a ese nodo
    result = []

    for vertice in Grafo:
        dist[vertice] = float("inf") #cota superior
        prev[vertice] = None #inicializamos cada valor de cada nodo
    dist[salida] = 0 #no hemos salido del vertice inicial

    Q = [vertice for vertice in Grafo] #se copia el diccionario grafo

    while Q:
        u = min(Q, key=dist.get)
        Q.remove(u)
        result.append(u)

        for vecino in Grafo[u]:
            if vecino in Q and dist[vecino] > dist[u] + Grafo[u][vecino]:
                dist[vecino] = dist[u] + Grafo[u][vecino] #sumamos la distancia desde el nodo de origen
                prev[vecino] = u

    return result, dist, prev


grafo_antenas = {
    'RU': {'BTC':fc(0), 'KPAMS': fc(0),'KPAMS1': fc(0),'KPAMS2': fc(0), 'KPAMS3': fc(0)},
    'BTC': {'RU': fc(0), 'KPAMS': fc(0),'KPAMS1': fc(0),'KPAMS2': fc(0), 'KPAMS3': fc(0)},
    'KPAMS': {'KPAMS1': fc(1),'KPAMS2': fc(1), 'KPAMS3': fc(1),'RU': fc(0), 'BTC': fc(0), 'TFSC': fc(0)},
    'KPAMS1': {'KPAMS': fc(1), 'KPAMS2': fc(1), 'KPAMS3': fc(1), 'TFSC1': fc(0), 'TFSC2': fc(0),'BTC': fc(0),'RU':fc(0)},
    'KPAMS2': {'KPAMS': fc(1), 'KPAMS1': fc(1), 'KPAMS3': fc(1), 'TFSC1': fc(0), 'TFSC2': fc(0),'BTC': fc(0),'RU':fc(0)},
    'KPAMS3': {'KPAMS': fc(1), 'KPAMS3': fc(1), 'KPAMS2': fc(1), 'TFSC1': fc(0), 'TFSC2': fc(0),'BTC': fc(0),'RU':fc(0)},
    'TFSC': {'TFSC1': fc(1), 'TFSC2': fc(1),'KPAMS': fc(0), 'SE':fc(0)},
    'TFSC1' : {'TFSC': fc(1), 'TFSC2': fc(1), 'SE1':fc(0), 'SE2':fc(0), 'SE3':fc(0),'UC1': fc(0),'UC2': fc(0),'KPAMS1': fc(0),'KPAMS2': fc(0), 'KPAMS3': fc(0), 'B':fc(0)},
    'TFSC2' : {'TFSC': fc(1), 'TFSC1': fc(1), 'SE1':fc(0), 'SE2':fc(0), 'SE3':fc(0),'UC1': fc(0),'UC2': fc(0),'KPAMS1': fc(0),'KPAMS2': fc(0), 'KPAMS3': fc(0), 'B':fc(0)},
    'SE': {'TFSC': fc(0), 'UC': fc(0), 'SE1':fc(1), 'SE2':fc(1), 'SE3':fc(1)},
    'SE1': {'SE':fc(1),'SE2':fc(1),'SE3':fc(1), 'TFSC1': fc(0), 'TFSC2': fc(0),'UC1': fc(0),'UC2': fc(0), 'B':fc(0)},
    'SE2': {'SE':fc(1),'SE1':fc(1), 'SE3':fc(1), 'TFSC1': fc(0), 'TFSC2': fc(0),'UC1': fc(0),'UC2': fc(0), 'B':fc(0)},
    'SE3': {'SE':fc(1),'SE1':fc(1), 'SE2':fc(1), 'TFSC1': fc(0), 'TFSC2': fc(0),'UC1': fc(0),'UC2': fc(0), 'B':fc(0)},
    'B': {'SE1':fc(0), 'SE2':fc(0), 'SE3':fc(0), 'TFSC1': fc(0), 'TFSC2': fc(0), 'UC1': fc(0),'UC2': fc(0)},
    'UC': {'UC1':fc(1), 'UC2':fc(1),'SE': fc(0)},
    'UC1' : {'UC':fc(1), 'UC2':fc(1), 'SE1':fc(0), 'SE2':fc(0), 'SE3':fc(0),'TFSC1': fc(0), 'TFSC2': fc(0), 'B':fc(0)},
    'UC2' : {'UC':fc(1), 'UC1':fc(1), 'SE1':fc(0), 'SE2':fc(0), 'SE3':fc(0),'TFSC1': fc(0), 'TFSC2': fc(0), 'B':fc(0)}
}

claves = list(grafo_antenas.keys())

for v in claves:
  s, distancia, previos = dijkstra1(grafo_antenas, v)
  print(f"{s=}")
  print(f"{distancia=}")
  print(f"{previos=}")
```
